```python
def set_values(row, value):
    return value[row]
map_dictionary ={"Sumapaz": '20 - SUMAPAZ', 'Rafael Uribe Uribe':'18 - RAFAEL URIBE URIBE', 'Candelaria':'17 - CANDELARIA', 'Barrios Unidos':'12 - BARRIOS UNIDOS','Teusaquillo': '13 - TEUSAQUILLO','Puente Aranda':'16 - PUENTE ARANDA','Los Mártires': '14 - LOS MÁRTIRES','Usaquén': '01 - USAQUÉN','Chapinero': '02 - CHAPINERO','San Cristóbal': '04 - SAN CRISTÓBAL','Usme': '05 - USME','Ciudad Bolívar': '19 - CIUDAD BOLÍVAR','Bosa': '07 - BOSA','Kennedy': '08 - KENNEDY','Engativá': '10 - ENGATIVÁ','Suba': '11 - SUBA','Tunjuelito':'06 - TUNJUELITO','Santa Fe': '03 - SANTA FE','Antonio Nariño': '15 - ANTONIO NARIÑO','Fontibón': '09 - FONTIBÓN'}
  
delitoxlocalidad["Localidad"] = delitoxlocalidad['Nombre Localidad'].apply(set_values, args =(map_dictionary, ))
```

```python
delitoxlocalidad
```

```python
coor_localidad=delitoxlocalidad[["Código Localidad","Nombre Localidad","coord_x","coord_y","Localidad"]]
```

```python
coor_localidad
```

```python
delitosBogota2=delitosBogota.merge(coor_localidad, left_on='Localidad', right_on='Localidad')
```

```python
delitosBogota2
```

```python
delitosBogota2.info()
```

```python
delitosBogota2['coord_x'] = delitosBogota2['coord_x'].apply(lambda x: x.replace(',','.'))
delitosBogota2['coord_y'] = delitosBogota2['coord_y'].apply(lambda x: x.replace(',','.'))
```

```python
delitosBogota2['coord_x'] = delitosBogota2['coord_x'].astype(float)
delitosBogota2['coord_y'] = delitosBogota2['coord_y'].astype(float)
```

```python
delitosBogota2.to_csv("Delitos Bogota2.csv",sep=";")
```

```python
delitoxlocalidad.to_csv("Delitos Localidad.csv",sep=";")
```

```python
plt.figure(figsize=(25,15))
sns.catplot(x = "Rango del Dia", data = delitosBogota2, kind = "count", hue = "Año");
```

```python
df2 = pd.DataFrame(columns = ['Offenses'])
df2["Offenses"]=[each for each in delitosBogota2.Delito.unique()]
df2["Count"]=[len(delitosBogota2[delitosBogota2.Delito==each]) for each in df2.Offenses]
df2=df2.sort_values(by=['Count'],ascending=False)

plt.figure(figsize=(25,15))
sns.barplot(x=df2.Offenses.head(50), y=df2.Count.head(50))
plt.xticks(rotation= 90)
plt.xlabel('Offenses')
plt.ylabel('Count')
plt.show()
```
